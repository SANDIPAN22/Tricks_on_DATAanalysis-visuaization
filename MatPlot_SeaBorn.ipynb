# DATA VISUALIZATION USING MATPLOTLIB

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(1,10,100)
```

```python
x
```

```python
y=x**x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X LABEL')
plt.ylabel('Y LABEL')
plt.title("DEMO PLOTTINGS")
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)


plt.subplot(2,1,2)
plt.plot(y,x)
```

## subplots usage in oops mode

```python
x1=np.linspace(1,5,10)
fig,ax=plt.subplots(2,2)
c=1
for a in ax:
    for k in a:
        
        y1=x1**c
        k.set_title(f'Mutiplied with{c}')
        c+=2
        
        k.plot(x1,y1)
```

```python
x1=np.linspace(1,5,10)
fig,ax=plt.subplots(2,2,figsize=(10,10))
c=1
for a in ax:
    for k in a:
        
        y1=x1**c
        k.set_title(f'Mutiplied with{c}')
        c+=5
        
        k.plot(x1,y1)
```

```python
fx,ax1=plt.subplots(figsize=(5,5))
ax1.bar(x1,x1**2,label='SQUARED ')
ax1.bar(x1,x1**3, label='TRIPPLED',alpha=0.5)
ax1.legend()
```

```python
fx.savefig('plotme.png')
```

## LOADING TITANIC DATASET

```python
st=sns.load_dataset('titanic')
st.head()
```

## * Try to find out the age distribution

```python
pa=pd.DataFrame(st['age'])
pp=pa.dropna(axis=0)
pp.head()
```

```python
sns.distplot(pp,bins=50)
```

## * sex distribution

```python
sns.countplot('sex',data=st)
```

## aggregate of fare classified by sex

```python
sns.barplot(x='sex',y='fare',data=st)
```

```python
sns.boxplot(x='sex',y='fare',data=st)
```

```python
sns.violinplot(x='sex',y='fare',data=st,hue='survived',split=True)
```

## CLASS 

```python
sns.countplot(st['class'])
```

```python
sns.distplot(st['survived'])
```

```python
sns.distplot(st['survived'])
```

## joint of last two

```python
sns.jointplot(x='survived',y='fare',data=st)
```

## joint b/w fare and pclass

```python
sns.jointplot(x='fare',y='pclass',data=st)
```

```python
sns.jointplot(x='fare',y='pclass',data=st,kind='hex')
```

```python
af=pd.DataFrame((st['age'],st['fare'])).transpose()
```

```python
af=af.dropna()
```

```python
af.head()
```

```python
sns.jointplot(af['age'],af['fare'],kind='hex')
```

```python
sns.jointplot(af['age'],af['fare'],kind='reg')
```

```python
sns.jointplot(af['age'],af['fare'],kind='kde')
```

```python
##using TIPS dataset understand pairplot


sns.pairplot(sp,hue='sex')
```

## clas and age

```python
sns.stripplot(x=st['class'],y=st['age'])
```

#### to feel density in a better way

```python
sns.stripplot(x=st['class'],y=st['age'],jitter=True)
```

## catagorize survive or not 

```python
sns.stripplot(x=st['class'],y=st['age'],jitter=True,hue=st['survived'],split=True)
```
