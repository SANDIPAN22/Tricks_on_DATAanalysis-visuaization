# GROUP_BY in PANDAS

```python
import numpy as np
import pandas as pd

ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
   'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}


p1=pd.DataFrame(ipl_data)
p1
```

```python
p1.groupby('Team')
```

```python
byTeam=p1.groupby('Team')
```

```python
byTeam.describe()
```

```python
byTeam['Year'].max()
```

```python
byRank=p1.groupby('Rank')
```

```python
byRank['Team'].count()
```

## FIND OUT WHICH TEAM WON THE MATCHES MAXIMUM TIME

```python
bb1=byRank.get_group(1)
```

```python
bb1
```

```python
bb2=bb1.groupby('Team')
```

```python
bb2['Team'].agg([np.size])[:1:]
```

## Applying Multiple Aggregation Functions at Once
With grouped Series, you can also pass a list or dict of functions to do aggregation with, and generate DataFrame as output −

```python
bb3=p1.groupby('Team')
bb3['Rank'].agg([np.sum,np.mean,np.std])

# group by teams the sum , mean, std of their rank
```
