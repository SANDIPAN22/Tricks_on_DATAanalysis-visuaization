# PANDAS VISUALIZATION & IO
* bar or barh for bar plots
* hist for histogram
* box for boxplot
* 'area' for area plots
* 'scatter' for scatter plots
* Pie Chart
* Line Chart

```python
import numpy as np
import pandas as pd
```

```python
p1=pd.DataFrame(np.random.randint(100,size=(10,5)),index=pd.date_range('1/1/2020',periods=10),columns=['sensor1','sensor2','sensor3','sensor4','sensor5'])
```

```python
p1
```

```python
p1.plot()
```

```python
p1.plot.bar()
```

```python
p1.plot.bar(stacked=True)
```

```python
p1.plot.barh()
```

```python
p1.plot.barh(stacked=True)
```

```python
p1.plot.hist(bins=50)
```

```python
p1.plot.box()
```

```python
p1.plot.area()
```

```python
p2=pd.DataFrame([10,20,30,40],index=['a','b','c','d'])
```

```python
p2.plot.pie(subplots=True)
```

```python
p3=pd.DataFrame(np.random.normal(loc=5,scale=2,size=(50,5)),columns=['A','B','C','D','E'])
```

```python
p3
```

```python
p3.plot.scatter(x='A',y='B',c='red')
p3.plot.scatter(x='A',y='C',c='green')
```

```python
df = pd.DataFrame({'mass': [2.87, 5.97, 6.00],
                   'radius': [6051.8, 6378.1, 71492]},
                  index=['Venus', 'Earth', 'Jupiter'])

df
```

```python
plot = df.plot.pie(y='mass', figsize=(5, 5))
```

```python
plot = df.plot.pie(subplots=True, figsize=(16, 14))
```
